### Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
cols = ['age','bmi','children']
scalar = StandardScaler()

df_cleaned[cols] = scalar.fit_transform(df_cleaned[cols])
```

```python
df_cleaned.head()
```

```python
from scipy.stats import pearsonr
```

```python
columns = df_cleaned.columns
columns
```

```python
# ----------------------------------
# Pearson Correlation Calculation
# ----------------------------------

# List of features to check against target
selected_features = [
    'age', 'bmi', 'children', 'is_female', 'is_smoker',
    'region_northwest', 'region_southeast', 'region_southwest',
    'bmi_category_Normal', 'bmi_category_Overweight', 'bmi_category_Obese'
]

correlation = {
    feature: pearsonr(df_cleaned[feature], df_cleaned['charges'])[0]
    for feature in selected_features
}

df_correlation = pd.DataFrame(list(correlation.items()), columns=['Feature','Pearson Correlation'])
df_correlation
```

```python
df_correlation.sort_values(by='Pearson Correlation', ascending=False)
```

```python
cat_features = [
    'is_female', 'is_smoker',
    'region_northwest', 'region_southeast', 'region_southwest',
    'bmi_category_Normal', 'bmi_category_Overweight', 'bmi_category_Obese'
]
```

```python
from scipy.stats import chi2_contingency

alpha = 0.05

df_cleaned['charges_bin'] = pd.qcut(df_cleaned['charges'], q=4, labels=False)
chi2_results = {}

for col in cat_features:
    contingency = pd.crosstab(df_cleaned[col], df_cleaned['charges_bin'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
chi2_df
```

```python
final_df = df_cleaned[['age', 'is_female', 'bmi', 'children', 'is_smoker', 'charges','region_southeast','bmi_category_Obese']]
```

```python
final_df
```

### Model training

```python
from sklearn.model_selection import train_test_split
```

X = target feature,
y = feature that will compared with all

```python
X = final_df.drop('charges', axis=1)
y = final_df['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
# create model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train the model
model.fit(X_train,y_train)
```

```python
# performance
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
```

```python
adjusted_r2 = 1 - ((1-r2)*(n-1))/(n-p-1)
adjusted_r2
```
