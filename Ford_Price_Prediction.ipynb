### Feature Engineering and Extraction

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.head()
```

```python
cols = ['year','mileage','tax','mpg']
scalar = StandardScaler()
df[cols] = scalar.fit_transform(df[cols])
```

```python
df.head()
```

```python
from scipy.stats import pearsonr
```

```python
df.columns
```

```python
selected_features = ['year', 'mileage', 'tax', 'mpg', 'engineSize',
       'transmission_Manual', 'transmission_Semi-Auto', 'model_ C-MAX',
       'model_ EcoSport', 'model_ Edge', 'model_ Escort', 'model_ Fiesta',
       'model_ Focus', 'model_ Fusion', 'model_ Galaxy', 'model_ Grand C-MAX',
       'model_ Grand Tourneo Connect', 'model_ KA', 'model_ Ka+',
       'model_ Kuga', 'model_ Mondeo', 'model_ Mustang', 'model_ Puma',
       'model_ Ranger', 'model_ S-MAX', 'model_ Streetka',
       'model_ Tourneo Connect', 'model_ Tourneo Custom',
       'model_ Transit Tourneo', 'model_Focus', 'fuelType_Electric',
       'fuelType_Hybrid', 'fuelType_Other', 'fuelType_Petrol']

correlations = {
    feature: pearsonr(df[feature], df['price'])[0]
    for feature in selected_features
}

correlation_df = pd.DataFrame(list(correlations.items()), columns=['Feature','Pearson Correlation Coefficient'])
```

```python
correlation_df
```

```python
df.drop(df['model_ C-MAX'],inplace=True)
```

```python
df
```

```python
df.drop('model_ EcoSport',axis=1,inplace=True)
```

```python
df.head()
```

```python
low_corr_features = correlation_df.loc[correlation_df['Pearson Correlation Coefficient'].abs() < 0.10, 'Feature'].tolist()
```

```python
low_corr_features
```

```python
df.drop(columns=low_corr_features, inplace=True, errors='ignore')
```

```python
df.shape
```

```python
df
```

### Machine Model

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

### performance

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_pred,y_test)
```
